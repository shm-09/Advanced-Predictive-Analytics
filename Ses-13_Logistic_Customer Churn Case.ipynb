```python
import pandas as pd
data = pd.read_excel("Datasets//Customer Churn.xlsx", sheet_name = "Case Data")
display(data)
```

```python
# Defining variables...
X = pd.DataFrame(data[['Customer Age (in months)', 'CHI Score Month 0', 'CHI Score 0-1', 'Support Cases Month 0', 'Support Cases 0-1', 'SP Month 0',  'SP 0-1',  'Logins 0-1', 'Blog Articles 0-1', 'Views 0-1', ' Days Since Last Login 0-1']])
y = pd.DataFrame(data[['Churn (1 = Yes, 0 = No)']])
```

```python
import statsmodels.api as sm

X = sm.add_constant(X)

model = sm.Logit(y, X)
result = model.fit(method='newton')

#### Evaluate the model#########

result.summary()
result.summary2()
```

#### Model Fitting using the Age Segments categories...
